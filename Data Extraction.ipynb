### SQLite database file

```python
sql = SQLite("sqlite:///data\\chinook.db")

df_left = pd.read_sql_query(
"""
select c.FirstName, c.LastName, i.Total
from 
    invoices i
    left join customers c
    on 
        i.CustomerId = c.CustomerId
    order by i.Total DESC
""", sql)

print(df_left['FirstName'][0], ' ', df_left['LastName'][0], 'has spend the most - $', df_left['Total'][0],' on invoices.',sep='')
```
